#1. Zadatak
###Konstruirajte neki neusmjereni netežinski graf sa 10 do 20 vrhova i reprezentirajte ga matricom susjedstva (ne put, ciklus, potpuni itd,...) u .txt file.

```python
import networkx as nx  #učitavanje biblioteke koja omogućuje manipulaciju i analizu grafova
import numpy as np     #učitavanje biblioteke za rad s višedimenzionalnim nizovima i matricama
```

```python
br_cvorova = 7
br_bridova = 10
G = nx.gnm_random_graph(br_cvorova, br_bridova)   #generiramo neusmjereni graf
G.add_edge(5, 5)
```

```python
matrica_susjedstva = nx.convert_matrix.to_numpy_array(G)   #pretvaramo graf u matricu susjedstva
```

```python
# Spremanje matrice susjedstva u .txt datoteku
with open('matrica_susjedstva.txt', 'w') as file:
    np.savetxt(file, matrica_susjedstva, fmt='%d')
```

#2. Zadatak
###Prikažite ga grafički u nekom alatu.

```python
import matplotlib.pyplot as plt    #učitavanje biblioteke za crtanje
```

```python
nx.draw(G, with_labels=True, node_size=700, node_color='pink', font_size=10)
plt.title('Prikaz neusmjerenog grafa')
plt.show()
```

#3. Zadatak
###Ispišite listu susjeda za taj graf, listu bridova i niz stupnjeva te broj bridova,prosječni stupanj i gustoću.


```python
# Ispis liste susjeda za svaki čvor
print("Lista susjeda za svaki čvor:")
for node in G.nodes():
    print(f"{node}: {list(G.neighbors(node))}")
```

```python
print("\nLista bridova:")
print(list(G.edges()))
```

```python
print("\nNiz stupnjeva:")
niz_stupnjeva=[degree for node, degree in G.degree()]
print(niz_stupnjeva)
```

```python
print("\nBroj bridova:")
print(G.number_of_edges())
```

```python
print("\nProsječni stupanj:")
print(sum(niz_stupnjeva) / br_cvorova)
```

```python
print(f"\nGustoća:{nx.density(G)}")
```

#4. Zadatak
###Nacrtajte histogram za distribuciju stupnjeva.

```python
degree_hist = nx.degree_histogram(G)   #vraća niz kojemu je i-ti element jednak broju čvorova u grafu koji imaju stupanj i

# Crtanje histograma
plt.bar(range(len(degree_hist)), degree_hist, color='pink')
plt.xlabel('Stupanj čvora')
plt.ylabel('Broj čvorova s tim stupnjem')
plt.title('Histogram distribucije stupnjeva')
plt.show()
```

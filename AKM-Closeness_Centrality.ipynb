Za input nekog vrha, ispisati tablicu udaljenosti od tog vrha do svih
ostalih i tablicu koliko vrhova je na kojoj udaljenosti.

```python
def bfs_distances(graph, source):
    # Inicijalizacija tablica udaljenosti i brojača
    distances = {}
    nodes_count = {}

    # Postavljanje početnog čvora na udaljenost 0 te postavljanje čvorova s udaljenosti 0 na 1
    distances[source] = 0
    nodes_count[0] = 1

    # Inicijalizacija BFS reda s početnim čvorom
    queue = [source]

    while queue:
        current_node = queue.pop(0)

        # Prolazak kroz susjedne čvorove
        for neighbor in graph.neighbors(current_node):
            # Ako susjedni čvor nije posjećen
            if neighbor not in distances:
                # Dodajemo ga u red
                queue.append(neighbor)
                # Postavljamo udaljenost do susjednog čvora
                distances[neighbor] = distances[current_node] + 1
                # Povećavamo brojač za odgovarajuću udaljenost
                if distances[neighbor] in nodes_count:
                    nodes_count[distances[neighbor]] += 1
                else:
                    nodes_count[distances[neighbor]] = 1

    return distances, nodes_count

# Input vrha
input_vertex = 0

# Računanje tablica udaljenosti i brojača
distances, nodes_count = bfs_distances(G, input_vertex)

# Ispis tablice udaljenosti od input vrha do svih ostalih vrhova
print("Udaljenost od vrha", input_vertex)
for node, distance in distances.items():
    print("Do vrha", node, ":", distance)

# Ispis tablice koliko vrhova ima na kojoj udaljenosti
print("\nBroj vrhova na svakoj udaljenosti")
for distance, count in nodes_count.items():
    print("Udaljenost", distance, ":", count)
```

Za input 2 vrha ispisati kolika je njihova udaljenost.

```python
def bfs_distance(graph, source, target):
    # Inicijalizacija BFS reda s početnim čvorom i udaljenošću 1, te skupa koji će sadržavat početne čvorove
    queue = [(source, 0)]
    visited = set()

    # BFS pretraga
    while queue:
        current_node, distance = queue.pop(0)  #uzimamo čvor iz BFS reda i njegove udaljenosti iz početnog čvora

        # Ako je trenutni čvor zapravo ciljni čvor, vraćamo udaljenost
        if current_node == target:
            return distance

        # Dodajemo trenutni čvor u skup posjećenih
        visited.add(current_node)

        # Dodajemo susjedne čvorove u red ako nisu posjećeni s povećanjem udaljenosti za 1 u odnosu na trenutni čvor
        for neighbor in graph.neighbors(current_node):
            if neighbor not in visited:
                queue.append((neighbor, distance + 1))

    # Ako ciljni čvor nije pronađen, vraćamo -1
    return -1

# Input dva vrha
input_vertex1 = 25
input_vertex2 = 11

# Računanje udaljenosti između dva vrha
distance = bfs_distance(G, input_vertex1, input_vertex2)

# Ispis udaljenosti
print("Udaljenost između vrhova", input_vertex1, "i", input_vertex2, "je:", distance)
```

Za input 2 vrha ispisati najkraće puteve između njih, npr.
input: vrhovi 4 i 12; output: najkraći putevi su: 4-2-5-12 i 4-10-5-12

```python
from collections import deque   #koristi se za manipuliranje redom čvorova u BFS algoritmu
def bfs_shortest_paths(graph, source, target):
    # Inicijalizacija BFS reda i rječnika posjećenih čvorova
    queue = deque([(source, [source])])  #Inicijalizacija BFS reda queue s početnim čvorom i putem koji sadrži samo početni čvor. Put se predstavlja kao lista čvorova.
    shortest_paths = []

    # Duljina najkraćih pronađenih puteva
    min_path_length = float('inf')   #postavljamo ju na beskonačno,a zatim ćemo pomoću nje pamtit najkraći put

    # BFS pretraga
    while queue:
        current_node, path = queue.popleft()

        # Ako smo dosegli ciljni čvor s minimalnom duljinom, dodajemo pronađeni put u listu najkraćih puteva
        if current_node == target and len(path) <= min_path_length:
            min_path_length = len(path)
            shortest_paths.append(path)
        # Ako smo dosegli ciljni čvor, a dužina puta prema njemu je veća od minimalne, prekidamo pretragu
        elif current_node == target:
            break

        # Dodajemo susjedne čvorove u red ako nisu posjećeni
        for neighbor in graph.neighbors(current_node):
            if neighbor not in path:
                queue.append((neighbor, path + [neighbor]))

    return shortest_paths

# Input dva vrha
input_vertex1 = 25
input_vertex2 = 28

# Računanje najkraćih puteva između dva vrha
shortest_paths = bfs_shortest_paths(G, input_vertex1, input_vertex2)

# Ispis najkraćih puteva
print("Najkraći putevi između vrhova", input_vertex1, "i", input_vertex2, "su:")
for i, path in enumerate(shortest_paths, 1):
    print(f"Putevi {i}: {'-'.join(map(str, path))}")
```

Učitajmo mrežu dupina:

```python
def load_graph_from_gml(filename):
    G = nx.Graph()
    with open(filename, "r") as f:
        graph_data = f.read()
    graph_data = graph_data.split("node")[1:]  # Uklanjanje početka koji sadrži "graph" i "directed" informacije

    # Dodavanje čvorova u graf
    for node_data in graph_data:
        node_id = int(node_data.split("id ")[1].split("\n")[0])
        node_label = node_data.split("label ")[1].split("\n")[0].replace('"', '')
        G.add_node(node_id, label=node_label)

    # Dodavanje bridova u graf
    edges_data = graph_data[-1].split("edge")[1:]
    for edge_data in edges_data:
        source = int(edge_data.split("source ")[1].split("\n")[0])
        target = int(edge_data.split("target ")[1].split("\n")[0])
        G.add_edge(source, target)

    return G

# Naziv GML datoteke
filename = "dolphins.gml"

# Učitavanje grafa
dupini = load_graph_from_gml(filename)

# Crtanje grafa
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(dupini)  # Postavljanje položaja čvorova
nx.draw(dupini, pos, with_labels=True, node_size=700, node_color="skyblue", font_size=10, font_weight="bold")  # Crtanje grafa
plt.title("Mreža dupina")  # Dodavanje naslova
plt.show()  # Prikaz grafa
```

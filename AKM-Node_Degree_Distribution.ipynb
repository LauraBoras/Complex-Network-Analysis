#**ZADATAK 1**
*Po ER modelu B, formirajte tri tipa random grafa s N=10000, za ùëùùëé = 0.00008, ùëùùëè = 0.0006 i ùëùùëê = 0.0018.
Za svaki od grafova ispi≈°ite distribuciju stupnjeva te nacrtajte histograme.
Izraƒçunajte i ispi≈°ite prosjeƒçni stupanj.*

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# Parametri modela B
N = 10000
pa = 0.00008
pb = 0.0006
pc = 0.0018

# Funkcija za generiranje grafa i crtanje histograma
def generate_and_plot_graph(p, label, color):
    # Generiranje grafa
    G = nx.erdos_renyi_graph(N, p)

    # Izraƒçun distribucije stupnjeva
    degrees = [deg for node, deg in G.degree()]

    # Ispis distribucije stupnjeva
    print(f"Distribucija stupnjeva za graf {label}:")
    print(nx.degree_histogram(G))

    # Izraƒçun prosjeƒçnog stupnja
    avg_degree = np.mean(degrees)
    print(f"Prosjeƒçni stupanj za graf {label}: {avg_degree}")

    # Crtanje histograma
    plt.hist(degrees, bins=range(min(degrees), max(degrees) + 2, 1), alpha=0.7, label=label, color=color)

# Generiranje i prikaz grafova
generate_and_plot_graph(pa, "A", "pink")
generate_and_plot_graph(pb, "B", "purple")
generate_and_plot_graph(pc, "C", "lightblue")

# Postavljanje naslova i legendi
plt.title('Distribucija stupnjeva za tri tipa random grafa')
plt.xlabel('Stupanj')
plt.ylabel('Broj ƒçvorova')
plt.legend()

# Prikaz histograma
plt.show()
```

#**ZADATAK 2**
*Za mre≈æu Condensed matter coauthorship (16726 vrhova) ispi≈°ite distribuciju stupnjeva, nacrtajte histogram i histogram po logaritamskoj skali. Odredite prosjeƒçni stupanj, koliki postotak vrhova pripada najveƒáoj komponenti te koja je prosjeƒçna udaljenost vrhova.*

```python
%cd /content/drive/MyDrive/Analiza kompleksnih mrezÃåa
```

```python
# Uƒçitavanje grafa iz datoteke
G = nx.read_edgelist("astro-ph.net")

# Broj ƒçvorova u poƒçetnom grafu
print("Broj ƒçvorova u poƒçetnom grafu:", G.number_of_nodes())

# Broj bridova u poƒçetnom grafu
print("Broj bridova u poƒçetnom grafu:", G.number_of_edges())
```

```python
# Ispis distribucije stupnjeva grafa
degree_distribution = nx.degree_histogram(G)
print("Distribucija stupnjeva grafa G:")
for i, count in enumerate(degree_distribution):
    if count > 0:
        print(f"Stupanj {i}: {count} ƒçvorova")

# Izvadi stupnjeve i broj ƒçvorova s tim stupnjem
degrees = range(len(degree_distribution))
node_count = degree_distribution
```

```python
# Nacrtaj histogram distribucije stupnjeva grafa
plt.figure(figsize=(10, 6))  # Prilagodi veliƒçinu figure
plt.bar(degrees, node_count, color='hotpink', alpha=0.7)
plt.xlabel('Stupanj vrha')
plt.ylabel('Broj vrhova s tim stupnjem')
plt.title('Distribucija stupnjeva grafa G')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()  # Prilagodi raspored elemenata
plt.show()
```

```python
# Nacrtaj histogram distribucije stupnjeva grafa s logaritamskom osnovom na x-osi
plt.figure(figsize=(10, 6))  # Prilagodi veliƒçinu figure
plt.bar(degrees, node_count, color='hotpink', alpha=0.7)
plt.xlabel('Stupanj vrha')
plt.ylabel('Broj vrhova s tim stupnjem')
plt.title('Distribucija stupnjeva grafa G (logaritamska osnova)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.yscale('log')  # Postavi logaritamsku skalu na y-os
plt.tight_layout()  # Prilagodi raspored elemenata
plt.show()
```

```python
# Izraƒçunaj prosjeƒçni stupanj grafa
average_degree = sum(dict(G.degree()).values()) / len(G)
print("Prosjeƒçni stupanj grafa G:", average_degree)
```

```python
# Izraƒçunaj najveƒáu komponentu
largest_component = max(nx.connected_components(G), key=len)
```

```python
# Izraƒçunaj postotak vrhova koji pripadaju najveƒáoj komponenti
percentage_largest_component = (len(largest_component) / len(G)) * 100
print("Postotak vrhova koji pripadaju najveƒáoj komponenti:", percentage_largest_component, "%")
```

```python
# Izraƒçunaj prosjeƒçnu udaljenost vrhova u najveƒáoj komponenti
average_shortest_path_length = nx.average_shortest_path_length(G.subgraph(largest_component))
print("Prosjeƒçna udaljenost vrhova u najveƒáoj komponenti:", average_shortest_path_length)
```
